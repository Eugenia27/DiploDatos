# Ejercicio 4:

Verificar que los proyectos que tienen nombres duplicados también tienen el hash de nombre duplicado


**Función que asigna a cada valor de la columna nombre, un string hexadecimal siempre que el valor no sea float, de ser float, devuelve un cateo a string** 

```
def hashit(val):
    if isinstance(val, float): 
        return str(val)
    return md5(val.encode('utf-8')).hexdigest()
```

```
duplicados['name'].apply(hashit).head(12)
```

**Observamos que para un mismo nombre se obtiene el mismo hash**

**Función que busca caracteres especiales o raros**

```
def weird(val):
    if isinstance(val, float): 
        return 0
    return bad.sequence_weirdness(val)
```

**Generamos una columna que contiene el valor de salida de la función anterior, es decir, el número de caracteres especiales por nombre o instancias raras en la cadena que le pasamos, para cada dataset**

```
kickstarter_2018_pandas['name_weirdness'] = kickstarter_2018_pandas['name'].apply(weird)
kickstarter_2018_pandas[kickstarter_2018_pandas['name_weirdness'] > 1][['name','name_weirdness']].head(6)
```

```
kickstarter_2018_iso['name_weirdness'] = kickstarter_2018_iso['name'].apply(weird)
kickstarter_2018_iso[kickstarter_2018_iso['name_weirdness'] > 1][['name','name_weirdness']].head(6)
```

**Obtenemos la cantidad de nombres con caracteres raros en cada dataset**

```
print('Con encoding iso-8859 se encuentran ****',len(kickstarter_2018_iso[kickstarter_2018_iso['name_weirdness'] > 1]),'**** nombres con caracteres especiales\n Mientras que sin encoding se tienen solo ****',len(kickstarter_2018_pandas[kickstarter_2018_pandas['name_weirdness'] > 1]),'****')
```

# Ejercicio 6:

Codificar currency con diccionario y con matriz de binarización


```
from sklearn import preprocessing
```

```
kickstarter_2018_pandas.describe(include=['O'])
```

```
column='currency'
# Codificamos las categorias con enteros que van de cero n donde n es el numero de categorias
# Creamos el objeto
le = preprocessing.LabelEncoder()
# Ajustamos el objeto le a las categorias de la columa currency
le.fit(kickstarter_2018_pandas[column])

```

**Primer mérodo de codificación a enteros a través de diccionario**

```
dict(zip(le.classes_,le.transform(le.classes_)))
```

```
# Aplicamos el objeto de codificacion y actualizamos los valores de currency
kickstarter_2018_pandas[column] = le.transform(kickstarter_2018_pandas[column]) 
kickstarter_2018_pandas.head()
```

```
# Recuperamos las categorias originales para aplicar el segundo metodo de codificacion
kickstarter_2018_pandas[column] = le.inverse_transform(kickstarter_2018_pandas[column]) 
```

```
from sklearn.preprocessing import LabelBinarizer

# Generamos un objeto de binarizacion
lb = LabelBinarizer()

# Aplicamos el objeto a la columna currency y obtenemos la matriz de ceros y unos
lb_results = lb.fit_transform(kickstarter_2018_pandas[column])
pd.DataFrame(lb_results, columns=((column + '_') + pd.Series(lb.classes_))).head(10)
```

# Ejercicio 8

Calcular los valores atípicos de 'usd_goal_real' y graficar los boxplots, con y sin estos valores por categoría


Calculamos primer y terce cuartil, rango intercuartil y los límites superior e inferior

```
q1 = kickstarter_2018_pandas.usd_goal_real.quantile(q=0.25) #primer cuartil
q3 = kickstarter_2018_pandas.usd_goal_real.quantile(q=0.75) #tercer cuartil

iq = q3-q1 # rango intercuartil
ls = q3+1.5*iq # limite superior 
li = q1-1.5*iq # limite inferior

print('Límite Inferior: ',round(li,2),'\nLímite Superior: ',round(ls,2))
```

**Dado que el límite inferior es negativo,Veamos cuál es el valor mínimo**

```
kickstarter_2018_pandas.usd_goal_real.min()
```

**Todos los valores son mayores que el límite inferior, por lo tanto, todos los outliers se encuentran por encima del límite superior, hacemos una tabla con tales valores**

```
valoresAtipicos = kickstarter_2018_pandas[kickstarter_2018_pandas.usd_goal_real>ls]
valoresAtipicos[['category','usd_goal_real']]
```

**Hacemos los diagramas de caja de usd_goal_real por categoría con y sin outliers. Los colores de las cajas representan a la categoría principal a la que pertenecen**

```
colores=[]
c=0
for i in range(16):
  c=c+i*i*1/100
  colores.append(cmap(c))
  
```

```
categorias=kickstarter_2018_pandas[['category','main_category']].drop_duplicates('category')
dicCatMCat = dict(zip(categorias.category.values, categorias.main_category.values))
dicMCatColor = dict(zip(categorias.main_category.unique(),colores))
```

```
custom_palette = {}
i=0
for q in set(kickstarter_2018_pandas.category):
    mc=dicCatMCat[q]
    custom_palette[q] = dicMCatColor[mc]
```

```
plt.figure(figsize=(125,8))
ax=sb.boxplot(x="category", y="usd_goal_real",data=kickstarter_2018_pandas,
              fliersize=5,notch=True,width=1, palette=custom_palette)
ax.set_xticklabels(ax.get_xticklabels(), rotation=75, ha="right")
plt.show()
```

```
plt.figure(figsize=(125,8))
ax=sb.boxplot(x="category", y="usd_goal_real",data=kickstarter_2018_pandas[kickstarter_2018_pandas.usd_goal_real<=ls],
              fliersize=0.001,notch=True,width=0.90, palette=custom_palette)
ax.set_xticklabels(ax.get_xticklabels(), rotation=75, ha="right")

markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colores]
plt.legend(markers, dicMCatColor.keys(), numpoints=1,bbox_to_anchor=(0.3, -0.4),fancybox=False,ncol=15)

plt.show()
```
